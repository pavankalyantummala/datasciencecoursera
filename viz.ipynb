```
"""In this program i am loading the 3-D nii image and making a slice of it
and storing it as sam.jpeg and again loading it to display"""

import nibabel as nib
import os
from nibabel.testing import data_path
#Loading the 3-D image
example_filename = os.path.join(data_path, 'T2.nii.gz')
img = nib.load("T2.nii.gz")
img = img.get_fdata()
slice_0 = img[:, 100, :]

#Store the slices in the file sam.jpeg
from PIL import Image
im = Image.fromarray(slice_0)
im = im.convert('L')
im.save("sam.jpeg")

#Get the image saved and display the image 
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('sam.jpeg')
imgplot = plt.imshow(img)
plt.show()
```

```
""" In this program I am making more slices of the image and displaying it here
 itself"""
example_filename = os.path.join(data_path, 'T2.nii.gz')
img = nib.load("T2.nii.gz")
img = img.get_fdata()
def show_slices(slices):
        """ Function to display row of image slices """
        fig, axes = plt.subplots(1, len(slices))
        for i, slice in enumerate(slices):
            axes[i].imshow(slice.T, cmap="gray", origin="lower")


slice_0 = img[:, 100, :]
slice_1 = img[100, :, :]
show_slices([slice_0,slice_1])
```
